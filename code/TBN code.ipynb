## Datasets

### Input matrix generator

```python
def InputGenes(n):
    output = list(itertools.product([0, 1], repeat=n))
    return output
```

### Logic gates definition

```python
def AND(x, y):
  results = np.where(x!=y, 0, x)
  return results
```

```python
def OR(x, y):
  results = np.maximum(x, y)
  return results
```

```python
def NOT(x):
  results = np.where(x==0, 1, 0)
  return results
```

### The logical rules of the FOS-GRN

```python
X=InputGenes(13)
Y=InputGenes(13)
```

```python
X = DataFrame(X, columns=['AG','AP1','AP2','AP3','EMF1','FT','FUL','LFY','PI','SEP','TFL1','UFO','WUS'])
Y = DataFrame(Y, columns=['AG','AP1','AP2','AP3','EMF1','FT','FUL','LFY','PI','SEP','TFL1','UFO','WUS'])
```

```python
a0=AND(NOT(X["EMF1"]), AND(NOT(X["AP2"]),NOT(X["TFL1"])))
a1=AND(NOT(X["EMF1"]), AND(NOT(X["AP1"]),X["LFY"]))
a2=AND(NOT(X["EMF1"]), AND(NOT(X["AP2"]),X["LFY"]))
a3=AND(NOT(X["EMF1"]),AND(NOT(X["TFL1"]),AND(X["LFY"],AND(X["AG"],X["SEP"]))))
a4=AND(NOT(X["EMF1"]), AND(X["LFY"],X["WUS"]))

Y["AG"]=OR(OR(OR(OR(a0,a1),a2),a3),a4)

b0=AND(NOT(X["AG"]),NOT(X["TFL1"]))
b1=AND(X["FT"],AND(X["LFY"],NOT(X["AG"])))
b2=AND(X["FT"],AND(NOT(X["AG"]),NOT(X["PI"])))
b3=AND(X["LFY"],AND(NOT(X["AG"]),NOT(X["PI"])))
b4=AND(X["FT"],AND(NOT(X["AG"]),NOT(X["AP3"])))
b5=AND(X["LFY"],AND(NOT(X["AG"]),NOT(X["AP3"])))

Y["AP1"]=OR(OR(OR(OR(OR(b0,b1),b2),b3),b4),b5)

Y["AP2"]=NOT(X["TFL1"])

Y["AP3"]=OR(AND(X["LFY"],X["UFO"]),AND(X["PI"],AND(X["SEP"],AND(X["AP3"],OR(X["AG"],X["AP1"])))))

Y["EMF1"]=NOT(X["LFY"])

Y["FT"]=NOT(X["EMF1"])

Y["FUL"]=AND(NOT(X["AP1"]),NOT(X["TFL1"]))

Y["LFY"]=OR(NOT(X["EMF1"]),NOT(X["TFL1"]))

Y["PI"]=OR(AND(X["LFY"],OR(X["AG"],X["AP3"])),AND(X["PI"],AND(X["SEP"],AND(X["AP3"],OR(X["AG"],X["AP1"])))))

Y["SEP"]=X["LFY"]

Y["TFL1"]=AND(NOT(X["AP1"]),AND(X["EMF1"],NOT(X["LFY"])))

Y["UFO"]=X["UFO"]

Y["WUS"]=AND(X["WUS"],(OR(NOT(X["AG"]),NOT(X["SEP"]))))
```

### Threshold Boolean network updating rule

```python
def evaluateModel(W,Th,Ig):
  res=np.zeros(len(Ig))
  aux=np.dot(W,Ig)
  aux1=aux-Th
  aux1=aux1.flatten()
  for j in range(0,len(Ig)):
    if aux1[j]>0:
      res[j]=1
    else:
      res[j]=0
  return (res)
```

```python
def evaluateModelN(W,Th,k):
  z=len(W)
  Ig=np.array(InputGenes(z))
  output=Ig
  for j in range(0,k):
    for i in range(0, len(Ig)):
      output[i,:]=evaluateModel(W,Th,output[i,:])
  return (output)
```
